```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\dav pe\Desktop\251090052618\Week 5\Cerelas.csv")

# Step A: Initial 5-number summary
print("Initial Summary:")
print(df.describe())

# Step B: Replace missing values (-1) with mean
for col in df.loc[:, "protein":"vitamins"].columns:
    mean_val = df[df[col] != -1][col].mean()
    df[col] = df[col].replace(-1, mean_val)

# Step C: Summary after missing value treatment
print("After Missing Value Treatment:")
print(df.describe())

# Step D: Replace noisy values with median
for col in df.loc[:, "protein":"vitamins"].columns:
    median_val = df[col].median()
    # Define noisy values (example: values beyond 3*std from mean)
    std = df[col].std()
    mean = df[col].mean()
    df.loc[(df[col] < mean - 3*std) | (df[col] > mean + 3*std), col] = median_val

# Step E: Summary after noisy value treatment
print("After Noisy Value Treatment:")
print(df.describe())
```
